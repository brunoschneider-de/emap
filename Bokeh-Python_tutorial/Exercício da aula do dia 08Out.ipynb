Pacotes:

```python
# pacotes
import pandas as pd
import os
from bokeh.palettes import Spectral11
from bokeh.plotting import figure, output_notebook, show, ColumnDataSource
from bokeh.models import HoverTool
import numpy as np
```

Alterando para o diretório de trabalho que contém os arquivos 'gapminder_fertility.csv' e 'gapminder_life_expectancy.csv.

```python
#os.chdir('diretório/que/contém/os/arquivos')
```

Carregando e tratando os dados.

```python
# carregando e tratando os dados os dados de fertilidade
fert = pd.read_csv('gapminder_fertility.csv', header=0)
fert.set_index('Country', inplace=True)
fert.dropna(axis = 0, how  = 'any', inplace=True)

# carregando e tratando os dados os dados de expectativa de vida
life = pd.read_csv('gapminder_life_expectancy.csv', header=0)
life.set_index('Country', inplace=True)
life.dropna(axis = 0, how  = 'any', inplace=True)
```

Países para os quais há dados:

```python
list(life.index)
```

Selecione três dos países da lista acima, e três cores referentes a cada país. Por exemplo:

```python
# selecionando 3 países
paises = ['Argentina', 'Italy', 'China']
#paises = ['Canada', 'Ireland', 'Japan']
#paises = ['France', 'Mexico', 'Malta']

# selecionando 3 cores
cores = ['darkblue', 'darkgreen', 'darkred']
#cores = ['black', 'orange', 'purple']
```

Fertilidade ao longo do tempo para os países selecionados:

```python
# output para Jupyter notebook
output_notebook()

# definindo um source para cada uma dos países selcionados.
source0 = ColumnDataSource(data = dict(
    fert = list(fert.loc[paises[0],:]),
    anos = list(fert.columns)
))

source1 = ColumnDataSource(data = dict(
    fert = list(fert.loc[paises[1],:]),
    anos = list(fert.columns)
))

source2 = ColumnDataSource(data = dict(
    fert = list(fert.loc[paises[2],:]),
    anos = list(fert.columns)
))

# criando um gráfico com título, e nome nos labels
p = figure(title="Fertility along time", width=900, height=400, x_axis_label = 'Year', y_axis_label = 'Fertility')

# adicionando um 'hover' para cada um das curvas
plot0 = p.line('anos', 'fert', line_width=2, color = cores[0], source = source0)
p.add_tools(HoverTool(renderers = [plot0], tooltips = [('Country', paises[0]), ('Year', '@anos'), ('Fertility', '@fert')]))

plot1 = p.line('anos', 'fert', line_width=2, color = cores[1], source = source1)
p.add_tools(HoverTool(renderers = [plot1], tooltips = [('Country', paises[1]), ('Year', '@anos'), ('Fertility', '@fert')]))

plot2 = p.line('anos', 'fert', line_width=2, color = cores[2], source = source2)
p.add_tools(HoverTool(renderers = [plot2], tooltips = [('Country', paises[2]), ('Year', '@anos'), ('Fertility', '@fert')]))

# apresentando os resultados
show(p)
```

Expectativa de vida ao longo do tempo para os países selecionados:

```python
# output para Jupyter notebook
output_notebook()

# definindo um source para cada uma dos países selcionados.
source0 = ColumnDataSource(data = dict(
    life = list(life.loc[paises[0],:]),
    anos = list(life.columns)
))

source1 = ColumnDataSource(data = dict(
    life = list(life.loc[paises[1],:]),
    anos = list(life.columns)
))

source2 = ColumnDataSource(data = dict(
    life = list(life.loc[paises[2],:]),
    anos = list(life.columns)
))

# criando um gráfico com título, e nome nos labels
p = figure(title="Life expectancy along time", width=900, height=400, x_axis_label = 'Year', y_axis_label = 'Life expectancy')

# adicionando um 'hover' para cada um das curvas
plot0 = p.line('anos', 'life', line_width=2, color = cores[0], source = source0, line_dash = 'dashed')
p.add_tools(HoverTool(renderers = [plot0], tooltips = [('Country', paises[0]), ('Year', '@anos'), ('Life expactancy', '@life')]))

plot1 = p.line('anos', 'life', line_width=2, color = cores[1], source = source1, line_dash = 'dashed')
p.add_tools(HoverTool(renderers = [plot1], tooltips = [('Country', paises[1]), ('Year', '@anos'), ('Life expactancy', '@life')]))

plot2 = p.line('anos', 'life', line_width=2, color = cores[2], source = source2, line_dash = 'dashed')
p.add_tools(HoverTool(renderers = [plot2], tooltips = [('Country', paises[2]), ('Year', '@anos'), ('Life expactancy', '@life')]))

# apresentando os resultados
show(p)
```
